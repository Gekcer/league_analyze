Найдем медиану каждого типа контроля вижна

```python
median_wards_placed = np.median(blue_wards['blueWardsPlaced'])
median_wards_destroyed = np.median(blue_wards['blueWardsDestroyed'])
```

теперь будем считать что если количество уничтоженных\поставленных вардов больше или равно медиане, то это хороший вижн (1),  
иначе - плохой (0)

```python
blue_wards['Good warding'] = blue_wards['blueWardsPlaced'].apply(lambda x: 1 if x >= median_wards_placed else 0)
blue_wards['Good destroy wards'] = blue_wards['blueWardsDestroyed'].apply(lambda x: 1 if x >= median_wards_destroyed else 0)
```

Посчитаем винрейт с хорошим вардингом и плохим вардингом. За хороший вардинг принимаем количество поставленных вардов больше  
медианы в победных играх. Аналогично для плохого вардинга

```python
high_warding_wr = blue_wards.loc[blue_wards['Good warding'] == 1]['blueWins'].sum() / blue_wards.loc[blue_wards['Good warding'] == 1].shape[0]
```

```python
low_warding_wr = blue_wards.loc[blue_wards['Good warding'] == 0]['blueWins'].sum() / blue_wards.loc[blue_wards['Good warding'] == 0].shape[0]
```

```python
print('Винрейт при хорошем вардинге = %r\nВинрейт при плохом вардинге = %r'%(high_warding_wr * 100,
                                                                             low_warding_wr * 100))
```

Видно, что винрейт немного повышается при хорошем вардинге  
Проверим аналагично для уничтожения вардов

```python
high_ward_clean_wr = blue_wards.loc[blue_wards['Good destroy wards'] == 1]['blueWins'].sum() / blue_wards.loc[blue_wards['Good destroy wards'] == 1].shape[0]
low_ward_clean_wr = blue_wards.loc[blue_wards['Good destroy wards'] == 0]['blueWins'].sum() / blue_wards.loc[blue_wards['Good destroy wards'] == 0].shape[0]
```

```python
print('Винрейт при хорошем контроле вардов = %r\nВинрейт при плохом контроле вардов = %r'%(high_ward_clean_wr * 100,
                                                                             low_ward_clean_wr * 100))
```

Как видно, хороший вардинг немного влияет на победу
